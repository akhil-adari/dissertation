```python
from scipy.optimize import minimize
import numpy as np

def function_p(x, p):
    a_z = np.sin(x[0])*np.cos(x[1])
    a_z_prime = np.sin(x[2])*np.cos(x[3])
    b_z = np.sin(x[4])*np.cos(x[5])
    b_z_prime = np.sin(x[6])*np.cos(x[7])
    a_perp = np.array([np.cos(x[1])*np.sin(x[0]),np.sin(x[1])*np.sin(x[0])])
    a_perp_prime = np.array([np.cos(x[3])*np.sin(x[2]),np.sin(x[3])*np.sin(x[2])])
    b_perp = np.array([np.cos(x[5])*np.sin(x[4]),np.sin(x[5])*np.sin(x[4])])
    b_perp_prime = np.array([np.cos(x[7])*np.sin(x[6]),np.sin(x[7])*np.sin(x[6])])
    return a_z*(b_z+b_z_prime)+a_z_prime*(b_z-b_z_prime)+p*(np.dot(a_perp,b_perp+b_perp_prime)+np.dot(a_perp_prime,b_perp-b_perp_prime))

```

```python
import matplotlib.pyplot as plt
initial_guess = np.random.rand(8)
p_values = np.linspace(0, 1, 100)
max_values = []
for p in p_values:
    result = minimize(lambda x: function_p(x,p), initial_guess, method='BFGS')
    max_values.append(np.abs(result.fun))

plt.plot(p_values, max_values)
plt.xlabel('p values')
plt.ylabel('Absolute value of maximum')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def A(theta):
    return np.cos(theta[0]) * (np.cos(theta[2]) + np.cos(theta[3])) + np.cos(theta[1]) * (np.cos(theta[2]) - np.cos(theta[3]))

def B(theta):
    return np.sin(theta[0]) * (np.sin(theta[2]) + np.sin(theta[3])) + np.sin(theta[1]) * (np.sin(theta[2]) - np.sin(theta[3]))

def F(p, theta):
    return np.abs(A(theta) + p * B(theta))

p_values = np.linspace(0, 1, num=100)
initial_guess = np.random.rand(4)
F_max_values = []
for p in p_values:
    result = minimize(lambda x: -F(p, x), initial_guess, method='BFGS')
    F_max_values.append(-result.fun)

plt.plot(p_values, F_max_values)
plt.xlabel("p values")
plt.ylabel("Absolute value for the maximum")
plt.title("p values vs Absolute value for the maximum")
plt.show()
```
